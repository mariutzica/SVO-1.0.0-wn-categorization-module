```python
import nltk
import ontology_category as oc
svo = oc.init_svo()

with open('tests/pitt_vars.txt', 'r') as f:
    pitt_vars = f.readlines()
pitt_vars = [x.strip().replace('_',' ') for x in pitt_vars] 
```

```python
for v in pitt_vars:
    print(f"variable: {v}")
    for word in v.split():
        print(f"\tword: {word}")
        defs = svo.what_is(word)
        cols = []
        if 'property' in defs.columns.values:
            cols.append('property')
        if 'phenomenon' in defs.columns.values:
            cols.append('phenomenon')
        if 'process' in defs.columns.values:
            cols.append('process')
        if 'state' in defs.columns.values:
            cols.append('state')
        if 'quantity' in defs.columns.values:
            cols.append('quantity')
        for index, row in defs.iterrows():
            print(f"\t\tdefinition: ({row['pos']}) {row['definition']}")
            if not 'state' in cols and ((row['pos']=='a') or (row['pos']=='s')):
                print(f"\t\t\tstate")
            for c in cols:
                if (row[c]=='yes') or (c=='state' and ((row['pos']=='a') or (row['pos']=='s'))):
                    print(f"\t\t\t{c}")
```
